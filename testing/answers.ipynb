```python

def keplerian_potential(M, r):
    """
    Function to define Keplerian potential.
    :param M: Mass of the larger body.
    :param r: Distance between the two bodies.
    :return: The Keplerian potential
    """
    # Gravitational constant (default is 1)
    G = 1
    # mass of smaller body (default is 1)
    m = 1
    return -G * M * m / r
print(keplerian_potential(100, 10))  # output: -10
print(keplerian_potential(200, 20))  # output: -10
import math


def angular_velocity(M, R):
    """
    Function to calculate the angular velocity of the guiding center.
    :param M: Mass of the central body
    :param R: The orbital radius
    :return: The angular velocity
    """
    G = 1  # Gravitational constant (default is 1)
    return math.sqrt(G * M / R ** 3)


# now we test our function
print(angular_velocity(100, 10))  # output will depend on the chosen value of G
print(angular_velocity(200, 20))  # output will depend on the chosen value of G
import math

def epicyclic_frequency(M, R, X):
    """
    Function to calculate the epicyclic frequency.
    :param M: Mass of the central body
    :param R: The orbital radius
    :param X: The maximum radial offset
    :return: The epicyclic frequency
    """
    G = 1  # Gravitational constant (default is 1)
    return math.sqrt((G * M / R**3) * (R**2 + 4*X**2))

# now we test our function

print(epicyclic_frequency(100, 10, 2))  # output will depend on the chosen value of G
print(epicyclic_frequency(200, 20, 4))  # output will depend on the chosen value of G
import math


def initialize_conditions(R, X, κ):
    """
    Function to initialize conditions for a body in an orbit.
    :param R: The orbital radius
    :param X: The maximum radial offset
    :return: A tuple containing the position and velocity vectors (r, v_r, v_theta)
    """
    r = R + X
    v_r = 0
    v_theta = - κ * X / 2
    return r, v_r, v_theta


# now we test our function
M, R, X = 100, 10, 2  # sample mass, orbital radius, and maximum radial offset
κ = math.sqrt((1 * M / R**3) * (R**2 + 4*X**2))  # calculating κ with G=1
print(initialize_conditions(R, X, κ))
import math

def star_position(t, R, X, Ωg, κ):
    """
    Function to calculate the position of a star.
    :param t: Time
    :param R: The orbital radius
    :param X: The maximum radial offset
    :param Ωg: The angular velocity of the guiding center.
    :param κ: The epicyclic frequency
    :return: A tuple containing the radial position and azimuth angle (r, θ)
    """
    r = R + X * math.cos(κ * t)
    θ = Ωg * t
    return r, θ

# now we test our function
M, R, X = 100, 10, 2  # sample mass, orbital radius, and maximum radial offset
κ = math.sqrt((1 * M / R**3) * (R**2 + 4*X**2))  # calculating κ with G=1
Ωg = math.sqrt(1 * M / R ** 3)  # calculating Ωg with G=1
print(star_position(15, R, X, Ωg, κ))  # outputs the position of the star at time 15 units
import matplotlib.pyplot as plt
import numpy as np

def plot_orbit(R, Ωg):
    """
    Function to plot the orbit of the guiding center.
    :param R: The orbital radius
    :param Ωg: The angular velocity of the guiding center.
    """
    # Generate an array of angles (0 to 2π)
    θ = np.linspace(0, 2 * np.pi * 3, 1000)  # multiply by 3 for three revolutions

    # Convert polar (R, θ) to cartesian (x, y) coordinates
    x = R * np.cos(θ)
    y = R * np.sin(θ)

    # Plot the orbit
    plt.figure(figsize=(7, 7))
    plt.plot(x, y)
    plt.title("Guiding Center's Orbit")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.gca().set_aspect("equal")
    plt.show()


# now we test our function
M, R = 100, 10  # sample mass and orbital radius
Ωg = np.sqrt(1 * M / R ** 3)  # calculating Ωg with G=1
plot_orbit(R, Ωg)
def plot_star_orbit(R, X, κ, Ωg):
    """
    Function to plot the orbit of a star.
    :param R: The orbital radius
    :param X: The maximum radial offset
    :param κ: The epicyclic frequency
    :param Ωg: The angular velocity of the guiding center.
    """
    # Generate an array of times (0 to the time of three full revolutions)
    t = np.linspace(0, 2 * np.pi * 3 / Ωg, 1000)
    
    # calculate the coordinates of the star at different times
    r = R + X * np.cos(κ * t)
    θ = Ωg * t
    
    # Convert polar (r, θ) to cartesian (x, y) coordinates
    x = r * np.cos(θ)
    y = r * np.sin(θ)

    # Plot the orbit
    plt.figure(figsize=(7, 7))
    plt.plot(y, x)  # swapping x and y to get clockwise rotation
    plt.title("Star's Orbit in Epicycle")
    plt.xlabel("Y")
    plt.ylabel("X")
    plt.grid(True)
    plt.gca().set_aspect("equal")
    plt.show()


# now we test our function
M, R, X = 100, 10, 2  # sample mass, orbital radius, and maximum radial offset
κ = math.sqrt((1 * M / R**3) * (R**2 + 4*X**2))  # calculating κ with G=1
Ωg = math.sqrt(1 * M / R ** 3)  # calculating Ωg with G=1
plot_star_orbit(R, X, κ, Ωg)
def plot_orbit(R, Ωg, color='blue'):
    θ = np.linspace(0, 2 * np.pi * 3, 1000)
    x = R * np.cos(θ)
    y = R * np.sin(θ)
    plt.plot(x, y, color=color)


def plot_star_orbit(R, X, κ, Ωg, color='red'):
    t = np.linspace(0, 2 * np.pi * 3 / Ωg, 1000)
    r = R + X * np.cos(κ * t)
    θ = Ωg * t
    x = r * np.cos(θ)
    y = r * np.sin(θ)
    plt.plot(y, x, color=color)  # swapping x and y to get clockwise rotation


# now we plot both orbits together
M, R, X = 100, 10, 2  # sample mass, orbital radius, and maximum radial offset
κ = math.sqrt((1 * M / R**3) * (R**2 + 4*X**2))  # calculating κ with G=1
Ωg = math.sqrt(1 * M / R ** 3)  # calculating Ωg with G=1

plt.figure(figsize=(7, 7))
plot_orbit(R, Ωg, color='blue')  # guiding center's circular orbit
plot_star_orbit(R, X, κ, Ωg, color='red')  # star's elliptical orbit
plt.title("Star's Orbit and Guiding Center's Orbit")
plt.xlabel("Cross-track")
plt.ylabel("Down-track")
plt.grid(True)
plt.gca().set_aspect("equal")
plt.show()
def plot_orbits_for_two_eccentricities(R, e1, e2, M, Ωg):
    # calculate maximum radial offset for each eccentricity
    X1 = e1 * R
    X2 = e2 * R

    # calculate epicyclic frequency for each maximum radial offset
    κ1 = math.sqrt((1 * M / R**3) * (R**2 + 4*X1**2)) 
    κ2 = math.sqrt((1 * M / R**3) * (R**2 + 4*X2**2))

    plt.figure(figsize=(7, 7))
    plot_orbit(R, Ωg, color='blue')  # guiding center's circular orbit
    plot_star_orbit(R, X1, κ1, Ωg, color='red')  # star's elliptical orbit for e1
    plot_star_orbit(R, X2, κ2, Ωg, color='green')  # star's elliptical orbit for e2
    plt.title("Star's Orbit and Guiding Center's Orbit for Two Eccentricities")
    plt.xlabel("Cross-track")
    plt.ylabel("Down-track")
    plt.grid(True)
    plt.gca().set_aspect("equal")
    plt.show()


# now we test our function
e1, e2 = 0.2, 0.4  # sample eccentricities
plot_orbits_for_two_eccentricities(R, e1, e2, M, Ωg)
def plot_orbits_for_various_eccentricities(R, e_values, M, Ωg):
    plt.figure(figsize=(7, 7))
    plot_orbit(R, Ωg, color='blue')  # guiding center's circular orbit

    # generate a different color for each eccentricity value
    colors = ['red', 'green', 'purple', 'orange', 'cyan']

    for e, color in zip(e_values, colors):
        X = e * R
        κ = math.sqrt((1 * M / R**3) * (R**2 + 4*X**2))
        plot_star_orbit(R, X, κ, Ωg, color=color)  # star's elliptical orbit 

    plt.title("Star's Orbit and Guiding Center's Orbit for Various Eccentricities")
    plt.xlabel("Cross-track")
    plt.ylabel("Down-track")
    plt.grid(True)
    plt.gca().set_aspect("equal")
    plt.show()


# now we test our function
eccentricities = [0.1, 0.3, 0.5, 0.7, 0.9]  # sample eccentricities
plot_orbits_for_various_eccentricities(R, eccentricities, M, Ωg)
```
